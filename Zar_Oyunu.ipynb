```python
# ZAR OYUNUNU, KİŞİYE KAZANDIN VEYA KAYBETTİN ŞEKLİNDE BİLGİ VEREREK SONUÇLANDIRALIM.

import turtle
import random

t1 = turtle.Turtle() #turtle sınıfından t1 isminde bir obje-nesne oluşturduk.

t1.speed(300) # bu objenin özelliklerine erişiyoruz.
t1.pensize(5)

zarAdet = int(input("Kaç defa zar atacaksın?"))

bir = iki = uc = dort = bes = alti = 0 # Aynı anda birden fazla değişken tanımlanabilir.

while True:
    tahmin = int(input("Sence hangi zar en fazla çıkacak?"))
    
    if not (tahmin>0 and tahmin<=6):
        print("Yanlış seçim. Böyle bir seçenek yok")
        continue
    else:
        
        for i in range(zarAdet):
            cikanZar = random.randint(1, 6)
            print(cikanZar)

            if cikanZar == 1:
                bir += 1           # bir = bir + 1 ile aynı anlamda.
            elif cikanZar ==2:
                iki += 1
            elif cikanZar ==3:
                uc += 1
            elif cikanZar ==4:
                dort += 1
            elif cikanZar ==5:
                bes += 1
            elif cikanZar ==6:
                alti += 1
     
    break
    
zarsayi=[1,2,3,4,5,6] # zarın üstünde olan rakamlar 1,2,3,4,5,6
mylist= []

for i in zarsayi:
    if i==1:
        mylist.append(bir) # append listeye eleman ekler.
    elif i==2:
        mylist.append(iki)
    elif i==3:
        mylist.append(uc)
    elif i==4:
        mylist.append(dort)
    elif i==5:
        mylist.append(bes)
    elif i==6:
        mylist.append(alti)
        
print("my list içindeki değerler", mylist)

enbuyuk = max(mylist)

print("Listedeki en büyük rakam", enbuyuk )

rakam = mylist.index(enbuyuk)

print(f"En büyük rakamın bulunduğu index {rakam} Bir ekleyerek hangi zarın en fazla çıktığını buluruz: {rakam+1}" )


if tahmin == (rakam+1):
    print(f"YOU WON..Your prediction is {tahmin}")
else:
    print(f"YOU LOST..Your prediction is {tahmin}, Computer guess is {rakam+1}")
    
```
